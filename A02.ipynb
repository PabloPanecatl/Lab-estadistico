```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_excel(r"C:\Users\pablo\OneDrive - ITESO\Semestre 5\Laboratorio de aprendizaje estad√≠stico\Boston Housing Dataset 1978.xlsx")
```

```python
df.isnull().sum()
```

```python
df2 = df.dropna()
df2.isnull().sum()
```

```python
y = df2['MEDV'].values
X_crim = df2['CRIM'].values.reshape(-1, 1)
X_zn = df2['ZN'].values.reshape(-1, 1)
X_indus = df2['INDUS'].values.reshape(-1, 1)
X_chas = df2['CHAS'].values.reshape(-1, 1)
X_nox = df2['NOX'].values.reshape(-1, 1)
X_rm = df2['RM'].values.reshape(-1, 1)
X_age = df2['AGE'].values.reshape(-1, 1)
X_dis = df2['DIS'].values.reshape(-1, 1)
X_rad = df2['RAD'].values.reshape(-1, 1)
X_tax = df2['TAX'].values.reshape(-1, 1)
X_ptratio = df2['PTRATIO'].values.reshape(-1, 1)
X_b = df2['B'].values.reshape(-1, 1)
X_lstat = df2['LSTAT'].values.reshape(-1, 1)
```

```python
n = len(y)
ones = np.ones((n, 1))
Xbig = np.hstack((ones, X_crim, X_zn, X_indus, X_chas, X_nox, X_rm, X_age, X_dis, X_rad, X_tax, X_ptratio, X_b, X_lstat))
```

```python
ols = sm.OLS(y, Xbig)
results = ols.fit()
results.summary()
```
