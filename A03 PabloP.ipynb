# A04 Aguas frescas

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import numpy as np
```

```python
datos = pd.read_excel("C:\\Users\\pablo\\OneDrive - ITESO\\Semestre 5\\Laboratorio de aprendizaje estadístico\\Aguas_frescas.xlsx")
datos
```

```python
#Ahora hacemos una regresión lineal múltiple usando sm.OLS
X = datos[['Kahlúa', 'Coco', 'Horchata', 'Mandarina', 'Tamaño', 'Hielo']]
y = datos['Calificación']
X = sm.add_constant(X)  # Agregar constante
modelo = sm.OLS(y, X).fit()
modelo.summary()
```

### Ahora hacemos lo mismo pero en vez de 0 a 1 es -1 a 1

```python
#Convertimos los booleanos de ser 0 a 1 a ser -1 a 1
X = datos[['Kahlúa', 'Coco', 'Horchata', 'Mandarina', 'Tamaño', 'Hielo']]
X = X.replace({0: -1, 1: 1})
#Y hacemos la regresión lineal múltiple con sm.OLS
X = sm.add_constant(X)  # Agregar constante
y = datos['Calificación']
modelo = sm.OLS(y, X).fit()
modelo.summary()
```

# Motor car trends

```python
carros = pd.read_excel("C:\\Users\\pablo\\OneDrive - ITESO\\Semestre 5\\Laboratorio de aprendizaje estadístico\\Motor Trend Car Road Tests.xlsx")
carros.columns
```

```python
X = carros[['cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']].values.reshape(-1, 10)
y = carros['mpg'].values.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.4, random_state=137)
# Escalamos los datos
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
ones = np.ones((X_train_scaled.shape[0], 1))
X_train_scaled = np.hstack((ones, X_train_scaled))  # Agregar columna de unos
# Creamos el modelo de regresión lineal con sm.OLS
X_train_scaled = sm.add_constant(X_train_scaled)  # Agregar constante
modelo = sm.OLS(y_train, X_train_scaled).fit()
modelo.summary()
```

### Ahora probamos el entrenado con la prueba

```python
#Ahora probamos el modelo con los datos test
X_test_scaled = sm.add_constant(X_test_scaled)  # Agregar constante
y_pred = modelo.predict(X_test_scaled)
r2_score(y_test, y_pred)
```
